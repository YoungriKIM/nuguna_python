```python
import pandas as pd
```

```python
root_path = 'D:/nuguna_python-1/data/reason_why_not_exercise/Report.xls'
not_exercise = pd.read_excel(root_path)
not_exercise.head()
```

```python
# 기간이라는 칼럼은 필요가 없으니 없애주자
not_exercise.drop(columns='기간', inplace=True)

# 27번 열부터는 지역 구별 데이터인데 이것들은 필요가 없으니 27번 열부터는 삭제하자
not_exercise.drop(index=range(22, 52), inplace=True)
not_exercise
```

```python
# 0번에 열인 '대분류'별로 데이터를 뽑아 그래프로 확인해보자
not_ex_gender = not_exercise[not_exercise['대분류']=='성별'].copy()
not_ex_gender
```

```python
## 여기서 잠깐 copy() 규칙을 들여다보자
test_a = [1,2,3,4]
print('origin_test_a:', test_a)

print('-'*10)
test_b = test_a
print('origin_test_b:',test_b)
test_c = test_a.copy()
print('origin_test_c:',test_c)

print('-'*10)
test_a[2] = 0
print('test_a:', test_a)
print('test_b:', test_b)
print('test_c:', test_c)
```

```python
# 다시 데이터로 돌아와서, 간결하게 보기 위해 '대분류' 열을 삭제하자
not_ex_gender.drop(columns='대분류', inplace=True)
not_ex_gender
```

```python
# 그래프를 그리기 쉽게 '분류' 열을 인덱스로 바꾸어주자
not_ex_gender.set_index('분류', inplace=True)
not_ex_gender
```

```python
# matplotlib을 사용해 그래프를 그리기 전 한글 설정을 해주자
import matplotlib.pyplot as plt

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

f_path = 'C:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font', family=font_name)
```

```python
f, ax = plt.subplots(1, 2, figsize=(16,8))

f.patch.set_facecolor('white') # 흰 배경

not_ex_gender['운동을 할 충분한 시간이 없어서'].plot.pie(explode=[0, 0.02], ax=ax[0], autopct='%1.1f%%')
ax[0].set_title('운동을 할 충분한 시간이 없어서')
ax[0].set_ylabel('')

not_ex_gender['함께 운동을 할 사람이 없어서'].plot.pie(explode=[0, 0.02], ax=ax[1], autopct='%1.1f%%')
ax[1].set_title('함께 운동을 할 사람이 없어서')
ax[1].set_ylabel('')

plt.show()
```

```python
# gender 말고 age로도 해보자
not_ex_age = not_exercise[not_exercise['대분류'] == '연령별'].copy()
not_ex_age.drop(columns='대분류', inplace=True)
not_ex_age.set_index('분류', inplace=True)
not_ex_age
```

```python
f, ax = plt.subplots(1, 5, figsize=(30,6))
f.patch.set_facecolor('white') # 흰 배경

explode_setting = [0.02,0.02,0.02,0.02,0.02,0.02]
sort = ['운동을 할 충분한 시간이 없어서', '함께 운동을 할 사람이 없어서', '운동을 할 만한 장소가 없어서', '운동을 싫어해서', '운동을 할 충분한 비용이 없어서']

for i in range(5):
    not_ex_age[sort[i]].plot.pie(explode = explode_setting, ax=ax[i], autopct='%1.1f%%')
    ax[i].set_title(sort[i])
    ax[i].set_ylabel('')
    
plt.show()
```
